```python
import sys
import torch
# Add a path to the module search path
sys.path.append('../')
from nodeood.data import DRData, DRDataset
from nodeood.data import ADSGenerator
from nodeood.datasets import Planetoid
from nodeood.methods import EERM
from nodeood.backbones import GCN
from nodeood.utils import set_random_seed
```

```python
set_random_seed(123)
```

```python
dataset_train = Planetoid('data', name='cora', mode='eerm-gat', split='train')
dataset_test = Planetoid('data', name='cora', mode='eerm-gat', split='test')
print(dataset_train)
print(dataset_test)
```

```python
data = dataset_train[0]
device = torch.device('cuda:0')
# Initialize model
gnn = GCN(in_channels=data.num_node_features, hidden_channels=16, out_channels=data.y.max().item()+1, num_layers=2, dropout=0.0).to(device)
# Initilize method
method = EERM(gnn=gnn, dataset=dataset_train, lr_g=0.01, lr_f=0.005, beta=1.0, epochs=200, K=3, T=1, S=5, weight_decay=1e-3, device=device)
```

```python
# Train model with early stopping
method.fit(dataset_train)
```

```python
# Evaluate model
method.test(dataset_test)
```
